<p>We going to use the class <strong>find_tasks</strong> and its method<strong> process_repos_for_tasks</strong>.</p>
<p>To create an instance of the class we need to provide the root directory of project running, and the directory of repositories.</p>

```python
root = os.getcwd()
root = root.replace('\\','/')
root
```

In my example the repositories are in a directory called <strong>repos</strong>.

```python
base = root+'/repos/'
```

```python
repoz = os.listdir(base)
len(repoz)
```

```python
find_tasks_instance = find_tasks(root,base)
```

<p>The method <strong>process_repos_for_tasks</strong> returns a dataframe containing the identified tasks and an integer of the total number of examined files</p>
<p>In the dataframe, in the column <strong>task_name</strong> the task name is stored and in column method_description the method body of the task is stored</p>
<p>The rest of the columns contain information about the file name, the repository it belongs and whether the file was placed under the directory <strong>tasks</strong> or <strong>roles</strong>.</p>
<p>Finally the files that PyYaml could not read are printed, which in total are 800 </p>

```python
df,sum_files = find_tasks_instance.process_repos_for_tasks(repoz)
```
