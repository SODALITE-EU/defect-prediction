```python
examined_repos = pd.DataFrame(columns=['repo_name','iac_percentage', 'contributors','commits'])
```

<h2>Test found ansible repositories for matching the criteria</h2>

```python
for repo in ansible_repos[767:]:
    try:
        repo_instance =  check_repo(token,repo,('.yml','.yaml'))
        iac_percentage = repo_instance.check_file_percentage()
        contributors = repo_instance.check_num_contributors()
        commits = repo_instance.check_commits()
        
        examined_repos = pd.concat([examined_repos,pd.DataFrame({'repo_name':repo,'iac_percentage':iac_percentage,
                                                                 'contributors':contributors,
                                                                 'commits':commits}
                                                        ,index=[0])],ignore_index = True)
        print(f'Repo {repo} added in list')
    except:
        print('Mistake found')
        continue
```

```python
passed_ansible_repos = examined_repos[(examined_repos.iac_percentage == True) & (examined_repos.commits == True) & (examined_repos.contributors == True)]
```

```python
passed_ansible_repos
```

```python
with open('passed_repos_ansible.pkl', 'wb') as output_file:
    pickle.dump(passed_ansible_repos, output_file)
```
